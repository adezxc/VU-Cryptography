```python
import numpy as np

alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

def letterize(numbers):
    return chr(numbers[1]) + chr(numbers[0])

def roundFunction(twoNumbers, k, f):
    r = twoNumbers[1]
    l = twoNumbers[0] ^ eval(f)
    return [r, l]
```
